```python
#import necessary libraries
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

#load the data frame
churn = pd.read_csv('churn_clean.csv')
churn.describe()
```

```python
#Ttest
Sample_A = churn[churn.Churn == "Yes"]['Email']
Sample_B = churn[churn.Churn == "No"]['Email']
t_result = stats.ttest_ind(Sample_A, Sample_B)
```

```python
#Distribution of Continuous Variables
monthly_charge_stats = churn['Email'].describe()
tenure_stats = churn['Contacts'].describe()

# Box plot for Email and Contacts
fig, axes = plt.subplots(1, 2, figsize=(10, 6))
axes[0].boxplot(churn['Email'])
axes[0].set_title('Box Plot - Emails')
axes[0].set_ylabel('Number of Emails')
axes[1].boxplot(churn['Contacts'])
axes[1].set_title('Box Plot - Contacts')
axes[1].set_ylabel('Number of Contacts')

# Display the summary statistics
print("Summary Statistics for Emails:\n", monthly_charge_stats)
print("\nSummary Statistics for Contacts:\n", tenure_stats)

# Show the box plots
plt.tight_layout()
plt.show()
```

```python
#Distribution of Categorical Variables
churn['Churn'].value_counts().plot.bar(rot=0)
plt.title('Churn')
plt.xlabel('Churn')
plt.ylabel('Count')
plt.show()

order = [1,2,3,4,5,6,7]
churn['Item7'].value_counts().loc[order].plot.bar(rot=0)
plt.title('Courteous Exchange')
plt.xlabel('Courteous Exchange Rating')
plt.ylabel('Count')
plt.show()
```

```python
#Distribution of Continuous Variables
churn.plot(x = 'Email', y = 'Contacts', kind = 'scatter')
plt.title("Emails and Contacts")
plt.show() 
```

```python
# Distribution of Catagorical Variables
plt.figure(figsize=(8, 6))
ax = sns.countplot(data=churn, x='Churn', hue='Item7')
plt.title('Bar Plot - Courteous Exchange by Churn')
plt.xlabel('Churn')
plt.ylabel('Count')
plt.legend(title='Courteous Exchange', loc='upper right')
plt.tight_layout()  # Adjusts the plot's layout
plt.show()
```
