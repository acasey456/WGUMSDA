```python
#import necessary libraries
import pandas as pd
#For array functionality
import numpy as np
#For preprocessing and analyzing the data
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

#load the data frame
df = pd.read_csv('teleco_market_basket.csv')
df.head()
```

```python
df.shape
df.info()
```

```python
#remove empty rows and replace null values to empty
df = df[df['Item01'].notna()]
df = df.fillna('')
df.shape
```

```python
#Create list of lists
transactions = df.values.tolist()
print(transactions[0])
```

```python
#feed list to transaction encoder
encoder = TransactionEncoder()
transactions = encoder.fit_transform(transactions)
df_encoded = pd.DataFrame(transactions, columns = encoder.columns_)
```

```python
#drop the empty column 
df_encoded = df_encoded.drop(columns='')
df_encoded.to_csv('Prepared_df_task3.csv', index = False)
```

```python
df_encoded
```

```python
# Apply the Apriori algorithm
frequent_itemsets = apriori(df_encoded, min_support = 0.02, use_colnames=True)
print(f'How many itemsets that met minimum support: {len(frequent_itemsets)}')
frequent_itemsets.head()
```

```python
# Generate the association rules with a minimum threshold of 1 for Lift
rules = association_rules(frequent_itemsets, num_itemsets = len(df_encoded),metric='lift', min_threshold=1)
print(f'How many rules exist: {(len(rules))}')
rules.head()
```

```python
# Sort by Support and print top 5
print(rules[['support','confidence','lift']])
```

```python
# print top 5
top_rules = rules.sort_values(by=['support', 'confidence', 'lift'], ascending=False)
top_rules.head()
```
